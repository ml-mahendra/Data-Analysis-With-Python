```python
# Data Normalization

#Uniform features wit diffrrent range
Col_names = ["length","width","height"]
multiple_columns = df1.loc[:,Col_names]
print(multiple_columns)
```

```python
# Normalizing the length using simple feature scale method

df1["length"] = df1["length"]/df1["length"].max()
df1["length"]
```

```python
# Normalizing the length using Min -Max method

df1["length"] = (df1["length"]-df1["length"].min())/(df1["length"].max() - df1["length"].min())
df1["length"]
```

```python
 # Normalizing the length using Z score method

df1["length"] = (df1["length"]-df1["length"].mean()) / df1["length"].std()
df1["length"]
```

```python
# Binning 

#we need 4 numbers as dividers that are equal distance apart. 
#First we use the numpy function “linspace” to return the array “bins” that contains 4 equally spaced numbers over 
#the specified interval of the price. 

bins = np.linspace(min(df1["price"]),max(df1["price"]), 4)

#We create a list “group_names “ that contains the different bin names. 
group_names = ['Low','Medium','High']

#We use the pandas function ”cut” to segment and sort the data values into bins.
df1["price-binned"] = pd.cut(df1["price"],bins,labels=group_names, include_lowest=True)

df1["price-binned"]
```
